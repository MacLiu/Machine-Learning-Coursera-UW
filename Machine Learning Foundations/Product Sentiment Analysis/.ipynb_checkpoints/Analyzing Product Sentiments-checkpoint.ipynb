```python
import graphlab as gl

#List of selected words to ustilize for better understanding of what the classifier is doing.
selected_words = ['awesome', 'great', 'fantastic', 'amazing', 'love', 'horrible', 'bad', 'terrible', 'awful', 'wow', 'hate']

products = gl.SFrame('amazon_baby.gl')
#display out data
products.head()
```

```python
products['word_count'] = gl.text_analytics.count_words(products['review'])
products['sentiment'] = products['rating'] >=4

word = ''

def count_words(word_count):
    if word in word_count:
        return word_count[word]
    else:
        return 0
    #endif
#enddef

for nWord in selected_words:
    word = nWord
    products[nWord] = products['word_count'].apply(count_words)
#endfor

products.head()
```

```python
train_data,test_data = products.random_split(.8, seed=0)

sentiment_model = gl.logistic_classifier.create(train_data,
                                                     target='sentiment',
                                                     features=['word_count'],
                                                     validation_set=test_data)

selected_words_model = gl.logistic_classifier.create(train_data,
                                                     target='sentiment',
                                                     features=selected_words,
                                                     validation_set=test_data)

selected_words_model.evaluate(test_data)

daiper_reviews = products[products['name'] == 'Baby Trend Diaper Champ']
daiper_reviews['predicted_sentiment'] = sentiment_model.predict(daiper_reviews, output_type='probability')
daiper_reviews['predicted_sentiment_selected'] = selected_words_model.predict(daiper_reviews, output_type='probability')

daiper_reviews = daiper_reviews.sort('predicted_sentiment', ascending=False)
daiper_reviews_selected = daiper_reviews.sort('predicted_sentiment_selected', ascending=False)

#compare the results of the predictions
daiper_reviews[0]['predicted_sentiment']
daiper_reviews_selected[0]['predicted_sentiment_selected']
```
